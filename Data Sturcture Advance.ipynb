```python
# Sure! Below is a Python code to check if a given string is a palindrome:

def is_palindrome(input_string):
    # Remove spaces and convert to lowercase for case-insensitive comparison
    cleaned_string = input_string.replace(" ", "").lower()
    
    # Compare the string with its reverse
    return cleaned_string == cleaned_string[::-1]

# Example usage
input_string = "A man a plan a canal Panama"
if is_palindrome(input_string):
    print(f"'{input_string}' is a palindrome.")
else:
    print(f"'{input_string}' is not a palindrome.")


#Explanation:
#The function is_palindrome first removes any spaces from the string and converts it to lowercase to handle case-insensitivity and spaces.
#It then compares the cleaned string with its reverse ([::-1]).
#If the string is equal to its reverse, it is a palindrome; otherwise, it's not.

#Example:
#For the input "A man a plan a canal Panama", the output will be:

# 'A man a plan a canal Panama' is a palindrome.

#Let me know if you'd like further details!
```

```python
# Certainly! Here’s a Python code to check if two given strings are anagrams of each other:
def are_anagrams(str1, str2):
    # Remove spaces and convert to lowercase for case-insensitive comparison
    str1 = str1.replace(" ", "").lower()
    str2 = str2.replace(" ", "").lower()
    
    # Check if sorted characters of both strings are the same
    return sorted(str1) == sorted(str2)

# Example usage
str1 = "listen"
str2 = "silent"
if are_anagrams(str1, str2):
    print(f"'{str1}' and '{str2}' are anagrams.")
else:
    print(f"'{str1}' and '{str2}' are not anagrams.")

#Explanation:
#The function are_anagrams first removes any spaces and converts both strings to lowercase to handle case-insensitivity and spaces.
#It then sorts the characters of both strings and compares them. If both sorted strings are equal, then the strings are anagrams of each other.
#sorted() converts a string into a sorted list of characters, which is then compared.

#Example:
#For the input strings "listen" and "silent", the output will be:
#'listen' and 'silent' are anagrams.

#Let me know if you have any more questions!
```

```python
#Certainly! Here's a Python code to find all occurrences of a given substring within another string:
def find_all_occurrences(main_string, substring):
    # Initialize a list to store the starting indices of all occurrences
    occurrences = []
    
    # Start searching from the beginning of the string
    start = 0
    while True:
        # Find the index of the next occurrence of the substring
        start = main_string.find(substring, start)
        
        # If no more occurrences are found, break the loop
        if start == -1:
            break
        
        # Add the found index to the list of occurrences
        occurrences.append(start)
        
        # Move the start position to search for the next occurrence
        start += 1
    
    return occurrences

# Example usage
main_string = "This is a test string. Testing is fun!"
substring = "test"
occurrences = find_all_occurrences(main_string, substring)

if occurrences:
    print(f"The substring '{substring}' occurs at the following positions: {occurrences}")
else:
    print(f"The substring '{substring}' was not found in the main string.")

#Explanation:
#The find_all_occurrences function uses a while loop to repeatedly find the next occurrence of the substring using the find() method.
#find() returns the index of the first occurrence of the substring, or -1 if the substring is not found.
#The start variable keeps track of where the search should begin to find the next occurrence, ensuring that all occurrences are found, even if they overlap.
#Each found index is added to the list occurrences, which is then returned.

#Example:
#For the input:
main_string = "This is a test string. Testing is fun!"
substring = "test"

#This shows the positions (indices) where "test" occurs in the main string.

#Let me know if you need further details!



```

```python
## question 6 >> Write a code to perform basic string compression using the counts of repeated characters
```

```python
#Certainly! Here's a Python code to perform basic string compression, where consecutive repeated characters are replaced by the character followed by the count of occurrences:

def compress_string(input_string):
    # If the string is empty, return it as is
    if not input_string:
        return input_string

    compressed_string = []
    count = 1  # Start with the first character

    # Iterate through the string, starting from the second character
    for i in range(1, len(input_string)):
        if input_string[i] == input_string[i - 1]:
            count += 1  # Increment the count if the current character matches the previous one
        else:
            compressed_string.append(input_string[i - 1] + str(count))  # Append the character and its count
            count = 1  # Reset count for the new character

    # Append the last character and its count
    compressed_string.append(input_string[-1] + str(count))

    # Join the list to form the compressed string
    return ''.join(compressed_string)

# Example usage
input_string = "aaabbbccccdaaa"
compressed_string = compress_string(input_string)
print(f"Original String: {input_string}")
print(f"Compressed String: {compressed_string}")

#Explanation:
#The function compress_string loops through the input string.
#It compares each character with the previous one to count how many times it repeats consecutively.
#For each group of repeated characters, it appends the character followed by the count to the compressed_string list.
#If there are no consecutive characters, it simply adds the character once.
#Finally, the list of compressed parts is joined into a single string and returned.

#Example:
#For the input string "aaabbbccccdaaa", the output will be:

#Original String: aaabbbccccdaaa
#Compressed String: a3b3c4d1a3

#This replaces repeated characters with their counts.

#Let me know if you'd like further clarification or adjustments!
```

```python
##question 7 >> Write a code to determine if a string has all unique characters

#Certainly! Below is a Python code to check if a string has all unique characters:

#Approach:
#Using a Set: The simplest and efficient way is to use a set. A set automatically handles duplicates, so if any character repeats, it will be detected.
#If the length of the set (which contains only unique characters) is equal to the length of the original string, then the string has all unique characters.

#Here’s the code:
def has_unique_characters(input_string):
    # Convert the string to a set to remove duplicate characters
    return len(input_string) == len(set(input_string))

# Example usage
input_string = "abcdefg"
if has_unique_characters(input_string):
    print(f"'{input_string}' has all unique characters.")
else:
    print(f"'{input_string}' does not have all unique characters.")

#Explanation:
#The function has_unique_characters converts the input string into a set. Since sets only store unique elements, any duplicate characters are discarded.
#It then compares the length of the original string with the length of the set. If they are the same, it means all characters were unique.

#Example:
#For the input string "abcdefg", the output will be:

#'abcdefg' has all unique characters.

#'aabbcc' does not have all unique characters.

#Time Complexity:
#O(n) where n is the length of the input string, since converting the string to a set and comparing the lengths both take linear time.
#Let me know if you'd like further details!
```

```python
##question 12 >> Implement a code to find the second largest number in a given list of integers

#Certainly! Below is a Python code to find the second largest number in a given list of integers:

#Code:
def second_largest(input_list):
    # Remove duplicates by converting the list to a set
    unique_numbers = set(input_list)

    # If there are less than two unique numbers, return None (no second largest)
    if len(unique_numbers) < 2:
        return None

    # Remove the largest element by popping the maximum
    unique_numbers.remove(max(unique_numbers))

    # The second largest number is now the maximum of the remaining set
    return max(unique_numbers)

# Example usage
input_list = [12, 45, 23, 45, 67, 89, 89, 34]
result = second_largest(input_list)

if result is not None:
    print(f"The second largest number is: {result}")
else:
    print("There is no second largest number.")

#Explanation:
#Remove Duplicates: First, we convert the list to a set to eliminate any duplicates. This ensures that we're working with unique numbers.
#Check Length: If there are fewer than two unique numbers, we return None because there's no second largest number.
#Find Largest: We use max() to find the largest number in the set, and then remove it using remove().
#Second Largest: After removing the largest, the second largest number is simply the new maximum of the remaining set.
#Example Output:
#For the input list [12, 45, 23, 45, 67, 89, 89, 34], the output will be:

#The second largest number is: 67


#Edge Cases:
#If the list has fewer than two unique numbers, the function will return None.
#Lists with only one unique number (e.g., [5, 5, 5]) will return None since there's no second largest number.

#Time Complexity:
#O(n): Where n is the length of the input list. The set conversion and finding the maximum both take linear time.
#Let me know if you need more details or modifications!
```

```python
## question 13>>>Create a code to count the occurrences of each element in a list and return a dictionary with elements as keys and their counts as values

#Certainly! Below is a Python code to count the occurrences of each element in a list and return a dictionary where the elements are the keys, and their counts are the values.

#Code:

def count_occurrences(input_list):
    # Create an empty dictionary to store the counts
    count_dict = {}
    
    # Iterate through the list
    for item in input_list:
        # If the item is already in the dictionary, increment its count
        if item in count_dict:
            count_dict[item] += 1
        else:
            # If the item is not in the dictionary, add it with a count of 1
            count_dict[item] = 1
    
    return count_dict

# Example usage
input_list = [1, 2, 3, 2, 1, 1, 3, 4]
result = count_occurrences(input_list)
print(f"Occurrences: {result}")


#Explanation:
#The function count_occurrences iterates through the input list.
#For each element, it checks if the element is already present in the dictionary count_dict.
#If the element is present, its count is incremented by 1.
#If the element is not present, it is added to the dictionary with an initial count of 1.
#Finally, the dictionary containing the elements and their counts is returned.

#Example Output:
#For the input list [1, 2, 3, 2, 1, 1, 3, 4], the output will be:

Occurrences: {1: 3, 2: 2, 3: 2, 4: 1}

#Time Complexity:
#O(n): Where n is the length of the input list. We only go through the list once, and dictionary operations (in, +=, and assignment) all take average constant time.

#Alternative Using collections.Counter:
#You can also use the built-in Counter class from the collections module, which is a more compact way to count occurrences.

from collections import Counter

def count_occurrences(input_list):
    return dict(Counter(input_list))

# Example usage
input_list = [1, 2, 3, 2, 1, 1, 3, 4]
result = count_occurrences(input_list)
print(f"Occurrences: {result}")

#This will give the same result, and is generally more efficient and concise when working with counting tasks.

#Let me know if you need further explanation or modifications!

```

```python
## question 14 >> Write a code to reverse a list in-place without using any built-in reverse functions

#Certainly! Below is a Python code to reverse a list in-place without using any built-in reverse functions:

#Code:

def reverse_list_inplace(input_list):
    # Initialize two pointers: one at the start and the other at the end
    start = 0
    end = len(input_list) - 1
    
    # Swap elements from both ends until the pointers meet in the middle
    while start < end:
        # Swap the elements at start and end
        input_list[start], input_list[end] = input_list[end], input_list[start]
        
        # Move the pointers towards the middle
        start += 1
        end -= 1
    
    return input_list

# Example usage
input_list = [1, 2, 3, 4, 5]
result = reverse_list_inplace(input_list)
print(f"Reversed list: {result}")

#Explanation:
#We use two pointers:
#One (start) starts from the beginning of the list.
#The other (end) starts from the end of the list.
#The while loop runs as long as start is less than end. During each iteration:
#We swap the elements at the start and end positions.
#After swapping, we move start one step forward and end one step backward, until they meet in the middle.
#This method modifies the list in-place, meaning it doesn't create a new list but changes the original list directly.

#Example Output:
#For the input list [1, 2, 3, 4, 5], the output will be:

#Reversed list: [5, 4, 3, 2, 1]

#Time Complexity:
#O(n): Where n is the length of the list. The loop runs for half of the list's length, so the time complexity is linear.
#Let me know if you'd like any further details or modifications!



```

```python
##question 15 >> Implement a code to find and remove duplicates from a list while preserving the original order of elements

#Certainly! Below is a Python code that removes duplicates from a list while preserving the original order of the elements.

#Code:

def remove_duplicates(input_list):
    # Create an empty set to track seen elements
    seen = set()
    
    # Create a new list to store the result without duplicates
    result = []
    
    # Iterate over the original list
    for item in input_list:
        # If the item hasn't been seen before, add it to the result and the set
        if item not in seen:
            result.append(item)
            seen.add(item)
    
    return result

# Example usage
input_list = [1, 2, 2, 3, 4, 4, 5, 3]
result = remove_duplicates(input_list)
print(f"List without duplicates: {result}")

#Explanation:
#We use a set seen to keep track of the elements that have already been encountered.
#We iterate through the original list. For each element:
#If it is not in the seen set, we append it to the result list and add it to the seen set.
#If the element has already been seen (i.e., it's in the seen set), we skip it.
#This approach ensures that only the first occurrence of each element is kept, and the order of the elements is preserved.

#Example Output:
#For the input list [1, 2, 2, 3, 4, 4, 5, 3], the output will be:

#List without duplicates: [1, 2, 3, 4, 5]

#Time Complexity:
#O(n): Where n is the length of the input list. Both checking membership in a set and adding an element to a set are O(1) operations on average, so the overall complexity is linear.
#Let me know if you'd like further clarifications or modifications!
```

```python
#question 16 >> Create a code to check if a given list is sorted (either in ascending or descending order) or not

#Sure! Here's a Python code that checks if a given list is sorted either in ascending order, descending order, or not sorted:

#Code:
def check_sorted(input_list):
    # If the list has fewer than 2 elements, it is considered sorted
    if len(input_list) < 2:
        return "Sorted"
    
    # Check if the list is sorted in ascending order
    is_ascending = all(input_list[i] <= input_list[i + 1] for i in range(len(input_list) - 1))
    
    # Check if the list is sorted in descending order
    is_descending = all(input_list[i] >= input_list[i + 1] for i in range(len(input_list) - 1))
    
    # Return the result based on the conditions
    if is_ascending:
        return "Ascending"
    elif is_descending:
        return "Descending"
    else:
        return "Not Sorted"

# Example usage
input_list = [1, 2, 3, 4, 5]
result = check_sorted(input_list)
print(f"List is sorted in: {result}")

input_list = [5, 4, 3, 2, 1]
result = check_sorted(input_list)
print(f"List is sorted in: {result}")

input_list = [1, 3, 2, 4, 5]
result = check_sorted(input_list)
print(f"List is sorted in: {result}")

#Explanation:
#Ascending Check: The code checks if each element in the list is less than or equal to the next element (input_list[i] <= input_list[i + 1]) using all(). This means the list is in ascending order if all the pairs satisfy this condition.

#Descending Check: Similarly, it checks if each element is greater than or equal to the next element (input_list[i] >= input_list[i + 1]) using all(). If true, the list is in descending order.

#If neither condition is satisfied, the list is considered "Not Sorted".

#For lists with fewer than 2 elements, we consider them sorted since there's nothing to compare.

#Example Output:
#For the input list [1, 2, 3, 4, 5], the output will be:

#List is sorted in: Ascending

#For the input list [5, 4, 3, 2, 1], the output will be:

#List is sorted in: Descending

#For the input list [1, 3, 2, 4, 5], the output will be:

#List is sorted in: Not Sorted

#Time Complexity:
#O(n), where n is the length of the input list. We go through the list once to check if it's sorted in ascending or descending order.
#Let me know if you need any further clarifications or modifications!
```

```python
## Question 17 >> Write a code to merge two sorted lists into a single sorted list

#Certainly! To merge two sorted lists into a single sorted list, we can use the two-pointer technique. This method involves comparing the elements of both lists from the beginning, and adding the smaller of the two elements to the result list at each step.

#Code:

def merge_sorted_lists(list1, list2):
    # Create two pointers for both lists
    i, j = 0, 0
    merged_list = []
    
    # Compare elements from both lists and merge them
    while i < len(list1) and j < len(list2):
        if list1[i] <= list2[j]:
            merged_list.append(list1[i])
            i += 1
        else:
            merged_list.append(list2[j])
            j += 1
    
    # If any elements are left in list1, append them
    while i < len(list1):
        merged_list.append(list1[i])
        i += 1
    
    # If any elements are left in list2, append them
    while j < len(list2):
        merged_list.append(list2[j])
        j += 1
    
    return merged_list

# Example usage
list1 = [1, 3, 5, 7]
list2 = [2, 4, 6, 8]
result = merge_sorted_lists(list1, list2)
print(f"Merged sorted list: {result}")

#Explanation:
#We initialize two pointers i and j to point to the current elements of list1 and list2, respectively.
#We compare the elements at list1[i] and list2[j]. The smaller element is added to the merged_list, and the corresponding pointer is incremented.
#Once we've gone through one of the lists, we append any remaining elements from the other list to the merged_list.

#Example Output:
#For the input lists list1 = [1, 3, 5, 7] and list2 = [2, 4, 6, 8], the output will be:

#Merged sorted list: [1, 2, 3, 4, 5, 6, 7, 8]

#Time Complexity:
#O(n + m), where n is the length of list1 and m is the length of list2. We process each element of both lists exactly once.
#This method is efficient because we are only iterating through the lists once, making it optimal for merging sorted lists.

#Let me know if you need further explanation or modifications!
```

```python
## question 19 >> Create a code to find the union of two lists without duplicates

#Certainly! To find the union of two lists without duplicates, we can use a set-based approach, which automatically removes duplicates and efficiently performs the union operation.

#Code:

def union_of_lists(list1, list2):
    # Convert both lists to sets and find the union
    return list(set(list1) | set(list2))

# Example usage
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]
result = union_of_lists(list1, list2)
print(f"Union of lists without duplicates: {result}")

#Explanation:
#Set Conversion: We convert both list1 and list2 into sets to eliminate any duplicates within each list.
#Union: We use the | operator to compute the union of the two sets, which combines all the unique elements from both lists.
#Convert back to List: The result is converted back to a list before returning, to match the desired output format.

#Example Output:
#For the input lists list1 = [1, 2, 3, 4, 5] and list2 = [4, 5, 6, 7, 8], the output will be:

#Union of lists without duplicates: [1, 2, 3, 4, 5, 6, 7, 8]


#Alternative Approach (Preserving Order):
#If you want to preserve the order of elements as they appear in the first list, you can iterate over both lists and add unique elements to the result list:

def union_of_lists_ordered(list1, list2):
    # Use a set to track elements already added to the result list
    result = []
    seen = set()

    # Add elements from list1
    for item in list1:
        if item not in seen:
            result.append(item)
            seen.add(item)

    # Add elements from list2
    for item in list2:
        if item not in seen:
            result.append(item)
            seen.add(item)
    
    return result

# Example usage
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]
result = union_of_lists_ordered(list1, list2)
print(f"Union of lists without duplicates (ordered): {result}")

#Example Output (Ordered):
#For the input lists list1 = [1, 2, 3, 4, 5] and list2 = [4, 5, 6, 7, 8], the output will be:
#Union of lists without duplicates (ordered): [1, 2, 3, 4, 5, 6, 7, 8]

#Time Complexity:
#Set-based approach: O(n + m), where n is the length of list1 and m is the length of list2. The set operations (conversion and union) are efficient and run in linear time.
#Ordered approach: O(n + m), where n and m are the lengths of list1 and list2. We process each element once, checking for membership in the seen set.
#Both methods are efficient, but the set-based approach is typically simpler and more concise. The ordered approach ensures the order of elements from the first list is maintained.

#Let me know if you have any questions or need further adjustments!
```

```python
## question 20 >> Write a code to shuffle a given list randomly without using any built-in shuffle functions

#Certainly! To shuffle a list randomly without using any built-in shuffle functions, we can implement a Fisher-Yates (Knuth) shuffle manually. This is an efficient algorithm that randomizes the order of elements in a list by swapping elements in place.

#Here’s how you can implement it:

#Code:

import random

def shuffle_list(input_list):
    # Iterate over the list from the last element to the second element
    for i in range(len(input_list) - 1, 0, -1):
        # Generate a random index from 0 to i
        j = random.randint(0, i)
        
        # Swap the elements at indices i and j
        input_list[i], input_list[j] = input_list[j], input_list[i]

    return input_list

# Example usage
input_list = [1, 2, 3, 4, 5]
shuffled_list = shuffle_list(input_list)
print(f"Shuffled list: {shuffled_list}")

#Explanation:
#Fisher-Yates Shuffle Algorithm:

#The Fisher-Yates shuffle works by iterating over the list from the last element to the second element. For each element at index i, it picks a random index j from 0 to i.
#The algorithm then swaps the elements at indices i and j. This ensures that all elements have an equal chance of being placed at any position in the list.
#Random Index Generation: The function random.randint(0, i) generates a random integer between 0 and i (inclusive), which is then used as the index to swap with the current element.

#Swapping: The two elements at indices i and j are swapped using the tuple swapping technique: input_list[i], input_list[j] = input_list[j], input_list[i].

#In-place Operation: This algorithm shuffles the list in place, meaning no additional space is required beyond temporary variables for the swap.

#Example Output:
#For the input list [1, 2, 3, 4, 5], the output will be something like:

#Shuffled list: [3, 5, 1, 4, 2]

#The order will vary each time you run the code because of the randomness.

#Time Complexity:
#O(n), where n is the length of the list. The algorithm performs a constant-time swap operation for each element of the list, so it runs in linear time.

#Space Complexity:
#O(1), since the shuffle is done in place with only a few variables used for temporary storage during the swap.

#This method is highly efficient and ensures that all possible permutations of the list are equally likely.

#Let me know if you need further clarifications or modifications!
```

```python
## Question 32 >> Write a code that takes two dictionaries as input and merges them into a single dictionary. If there are common keys, the values should be added together

#Here's a Python code that takes two dictionaries as input, merges them into a single dictionary, and if there are common keys, the values are added together:

def merge_dictionaries(dict1, dict2):
    # Create a new dictionary to store the merged result
    merged_dict = dict1.copy()  # Start with a copy of the first dictionary
    
    # Iterate through the second dictionary
    for key, value in dict2.items():
        # If the key is already in the merged dictionary, add the values
        if key in merged_dict:
            merged_dict[key] += value
        else:
            merged_dict[key] = value
    
    return merged_dict

# Example usage
dict1 = {'a': 1, 'b': 2, 'c': 3}
dict2 = {'b': 3, 'c': 4, 'd': 5}

merged_dict = merge_dictionaries(dict1, dict2)
print("Merged dictionary:", merged_dict)

#Explanation:
#merge_dictionaries function:

#We start by creating a copy of the first dictionary (dict1) to avoid modifying it directly.
#We then loop over each key-value pair in the second dictionary (dict2).
#If the key already exists in the merged dictionary, we add the corresponding values together.
#If the key does not exist in the merged dictionary, we simply add the key-value pair.

#Example usage:
#We define two dictionaries dict1 and dict2.
#The merge_dictionaries function is called to merge the two dictionaries.
#The result is printed.

#Example Output:

#Merged dictionary: {'a': 1, 'b': 5, 'c': 7, 'd': 5}


#In this example:
#
#The key 'b' is common in both dictionaries, so its values (2 and 3) are added together, resulting in 5.
#Similarly, for the key 'c', the values (3 and 4) are added together, resulting in 7.
#The keys 'a' and 'd' are unique, so their values remain as they are in the merged dictionary.
#This approach handles both unique and common keys correctly, adding values when necessary.
```

```python
##question 31>>  Write a code that takes a list of words as input and returns a dictionary where the keys are unique words and the values are the frequencies of those words in the input list

#Here's a Python code that takes a list of words as input and returns a dictionary where the keys are unique words and the values are the frequencies of those words in the input list:
def word_frequencies(word_list):
    # Create an empty dictionary to store word frequencies
    frequency_dict = {}
    
    # Iterate through each word in the list
    for word in word_list:
        # If the word is already in the dictionary, increment its count
        if word in frequency_dict:
            frequency_dict[word] += 1
        else:
            # If the word is not in the dictionary, add it with a count of 1
            frequency_dict[word] = 1
    
    return frequency_dict

# Example usage
words = input("Enter a list of words separated by spaces: ").split()
result = word_frequencies(words)
print("Word frequencies:", result)



#Explanation:
#word_frequencies function:
#This function takes a list of words (word_list) as input.
#It initializes an empty dictionary frequency_dict to store the frequency of each word.
#Then, it loops through each word in the input list:
#If the word is already a key in the dictionary, its value (the frequency) is incremented by 1.
#If the word is not in the dictionary, it is added with an initial frequency of 1.

#Input and output:
#The user is prompted to enter a list of words separated by spaces. This input is then split into a list using the split() method.
#The function word_frequencies() is called with the list of words, and the result is printed, which is a dictionary where keys are the unique words and values are their respective frequencies.

#Example Output:

#Enter a list of words separated by spaces: apple banana orange apple orange apple
#Word frequencies: {'apple': 3, 'banana': 1, 'orange': 2}

#Key Points:
#The function correctly handles counting the occurrences of each word in the list.
#It uses a dictionary to store each unique word as a key, and its count as the value.
#The user can input a list of words separated by spaces, and the code will calculate the frequency of each word.
#This approach is efficient for counting word frequencies and works well with input lists of any size.

```

```python
## Question 33>> Write a code to access a value in a nested dictionary. The function should take the dictionary and a list of keys as input, and return the corresponding value. If any of the keys do not exist in the dictionary, the function should return None

#Here's a Python function that takes a nested dictionary and a list of keys as input, and returns the corresponding value. If any of the keys do not exist in the dictionary, the function returns None.

def access_nested_value(nested_dict, keys):
    # Loop through the list of keys and access the corresponding values in the dictionary
    for key in keys:
        # If the key does not exist in the current level of the dictionary, return None
        if key not in nested_dict:
            return None
        # Move to the next level of the nested dictionary
        nested_dict = nested_dict[key]
    
    return nested_dict  # Return the value if all keys exist

# Example usage
nested_dict = {
    'a': {
        'b': {
            'c': 10,
            'd': 20
        },
        'e': 30
    },
    'f': 40
}

keys = ['a', 'b', 'c']
result = access_nested_value(nested_dict, keys)
print("Result:", result)  # Output: 10

# If a key does not exist
keys_invalid = ['a', 'b', 'z']
result_invalid = access_nested_value(nested_dict, keys_invalid)
print("Result for invalid keys:", result_invalid)  # Output: None

#Explanation:
#Function access_nested_value:
#This function takes two arguments:
#nested_dict: The nested dictionary where we want to access a value.
#keys: A list of keys that represent the path to the value.
#The function iterates over each key in the keys list:
#If the key is found in the current level of the dictionary, it accesses the next level by updating nested_dict to nested_dict[key].
#If at any point the key does not exist in the current dictionary, it returns None.

#Example usage:
#We have a nested_dict with multiple levels. The first test accesses the value of 'c' by following the keys ['a', 'b', 'c'], which returns 10.
#The second test demonstrates what happens when a key (in this case 'z') does not exist in the path. It returns None.

#Example Output:
#Result: 10
#Result for invalid keys: None

#Key Points:
#The function handles nested dictionaries of arbitrary depth and checks for the existence of keys at each level.
#It ensures that if any key does not exist, the function immediately returns None.
#This is a safe way to access deeply nested values without causing a KeyError.
```

```python
## Question 34 >> Write a code that takes a dictionary as input and returns a sorted version of it based on the values. You can choose whether to sort in ascending or descending order 

def sort_dict_by_value(input_dict, descending=False):
    # Sort the dictionary based on its values
    sorted_dict = dict(sorted(input_dict.items(), key=lambda item: item[1], reverse=descending))
    return sorted_dict

# Example usage
input_dict = {'apple': 10, 'banana': 5, 'orange': 15, 'grape': 8}

# Sort in ascending order (default)
sorted_dict_asc = sort_dict_by_value(input_dict)
print("Sorted dictionary (ascending):", sorted_dict_asc)

# Sort in descending order
sorted_dict_desc = sort_dict_by_value(input_dict, descending=True)
print("Sorted dictionary (descending):", sorted_dict_desc)

#Explanation:

#sort_dict_by_value function:
#This function takes two arguments:
#input_dict: The dictionary that needs to be sorted.
#descending: A boolean flag (default is False) that specifies whether the dictionary should be sorted in descending order. If True, the sorting will be in descending order; otherwise, it will be in ascending order.
#The function uses Python's built-in sorted() function to sort the dictionary items. The key=lambda item: item[1] part tells sorted() to use the values (item[1]) as the sorting criterion.
#The reverse=descending parameter is used to determine the sort order (ascending if False, descending if True).
#The result is converted back into a dictionary using dict().

#Example usage:
#We define a dictionary input_dict with some sample data.
#First, we call sort_dict_by_value to sort the dictionary in ascending order (default).
#Then, we call the same function with descending=True to sort the dictionary in descending order.

#Example Output:
#Sorted dictionary (ascending): {'banana': 5, 'grape': 8, 'apple': 10, 'orange': 15}
#Sorted dictionary (descending): {'orange': 15, 'apple': 10, 'grape': 8, 'banana': 5}

#Key Points:
#The dictionary is sorted based on the values, not the keys.
#The descending argument allows for flexible sorting in either ascending or descending order.
#Sorting a dictionary results in a list of tuples (key, value), which is then converted back into a dictionary.
#This approach works efficiently for sorting dictionaries of any size based on their values.
```

```python
## question 35 >> Write a code that inverts a dictionary, swapping keys and values. Ensure that the inverted dictionary correctly handles cases where multiple keys have the same value by storing the keys as a list in the inverted dictionary.

#To invert a dictionary, where we swap keys and values, we need to handle cases where multiple keys share the same value. Instead of storing the value as a single key, we store the keys as a list of keys associated with that value.

#Here’s the Python code to achieve this:

def invert_dict(input_dict):
    inverted_dict = {}
    
    # Loop through the input dictionary
    for key, value in input_dict.items():
        # If the value is already a key in the inverted dictionary, append the key to the list
        if value in inverted_dict:
            inverted_dict[value].append(key)
        else:
            # Otherwise, create a new entry with the value as the key and the key as a list
            inverted_dict[value] = [key]
    
    return inverted_dict

# Example usage
input_dict = {'a': 1, 'b': 2, 'c': 1, 'd': 3, 'e': 2}
inverted_dict = invert_dict(input_dict)
print("Inverted dictionary:", inverted_dict)

#Explanation:

#invert_dict function:
#This function takes a dictionary (input_dict) as input and returns a new dictionary where the keys and values are swapped.
#We iterate over each key-value pair in the input dictionary:
#If the value is already a key in the inverted dictionary, we append the current key to the list of keys that share this value.
#If the value is not in the inverted dictionary, we create a new entry with the value as the key and a list containing the current key as its value.

#Example usage:
#The dictionary input_dict has some repeated values (1 and 2).
#After inverting the dictionary, the values that appeared more than once are stored as lists of their respective keys.

#Example Output:

#Inverted dictionary: {1: ['a', 'c'], 2: ['b', 'e'], 3: ['d']}

#Key Points:
#Handling duplicates: If multiple keys have the same value, they are stored in a list under that value in the inverted dictionary.
#Flexible for any dictionary: This code works for dictionaries where values can be either unique or repeated.

#This solution handles all cases efficiently, ensuring that the inverted dictionary preserves the relationships between keys and values, even when values are not unique.
```
